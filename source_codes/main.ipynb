```python
def mean(data):
    return sum(data)/len(data)

def mode(data):
    dmax = data[0]
    for d in data:
        if data.count(d) > data.count(dmax):
            dmax = d
    return dmax

def median(data):
    data = sorted(data)
    n = len(data)
    if n % 2 == 1:
        return data[n//2]
    else:
        i = n//2
        return (data[i - 1] + data[i])/2
    
mean=mean(list(song_df['Total Streams']))
mode_=mode(list(song_df['Total Streams']))
median=median(list(song_df['Total Streams']))
print('Mean=',mean)
print('Median=',median)
print('Mode=',mode_)
```
